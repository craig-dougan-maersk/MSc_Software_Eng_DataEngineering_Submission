```python
import pandas as pd

dataset_2019_2023_actual = pd.read_csv('https://raw.githubusercontent.com/Green-Software-Foundation/real-time-cloud/refs/tags/v1.0/Cloud_Region_Metadata.csv')

dataset_2024_2025_estimate = pd.read_csv('https://raw.githubusercontent.com/Green-Software-Foundation/real-time-cloud/refs/tags/v1.0/Cloud_Region_Metadata_estimate.csv')

yearly_energy_dataset = pd.read_csv('https://storage.googleapis.com/emb-prod-bkt-publicdata/public-downloads/yearly_full_release_long_format.csv')

'''Check the two datasets have the same columns'''
#(https://www.geeksforgeeks.org/python/python-difference-two-lists/)
```

```python
'''check estimated dataset'''
dataset_2024_2025_estimate.head()
```

```python
def column_differences():
    unique_2019_2023 = list(set(dataset_2019_2023_actual.columns) - set(dataset_2024_2025_estimate.columns))
    unique_2024_2025 = list(set(dataset_2024_2025_estimate.columns) - set(dataset_2019_2023_actual.columns))
    return {
        "2019_2023" : {
            "unique_columns_list" : unique_2019_2023,
            "total_unique_columns" : len(unique_2019_2023),
            "total_columns" : len(dataset_2019_2023_actual.columns),
        },
        "2024_2025" : {
            "unique_columns_list" : unique_2024_2025,
            "total_unique_columns" : len(unique_2024_2025),
            "total_columns" : len(dataset_2024_2025_estimate.columns),
        },
        "combined" : {
            "unique_columns_list" : unique_2019_2023 + unique_2024_2025,
            "total_unique_columns" : len(unique_2019_2023 + unique_2024_2025)
        }
}
column_differences()
```

```python
dataset_2024_2025_estimate.drop(column_differences()["2024_2025"]["unique_columns_list"], axis=1, inplace=True)
dataset_2024_2025_estimate.head()
```
